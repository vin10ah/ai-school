# Matplotlib와 Pyplot 소개

Matplotlib는 파이썬에서 데이터를 시각화하는 데 널리 사용되는 라이브러리입니다. 다양한 종류의 차트와 그래프를 그릴 수 있으며, 데이터 분석과 과학 계산에서 유용하게 사용됩니다. Matplotlib는 매우 유연하며 사용자 정의가 가능한 강력한 API를 제공합니다.  

Pyplot는 Matplotlib의 서브패키지로, Matplotlib의 기능을 간편하게 사용할 수 있도록 인터페이스를 제공하는 모듈입니다. Pyplot를 이용하면 간단한 명령어로 복잡한 시각화를 손쉽게 만들 수 있습니다.  

Pyplot는 Matplotlib의 함수형 인터페이스로, MATLAB과 유사한 방식으로 작동합니다. 즉, 하나의 함수 호출로 그림을 만들고, 데이터를 플로팅하며, 라벨을 추가하는 등 다양한 시각화 작업을 수행할 수 있습니다.


```python
import matplotlib.pyplot as plt
```

### 무작위 데이터 생성

```python
import pandas as pd
import numpy as np

# 무작위 데이터 생성
np.random.seed(0)
arr = np.random.rand(20)
data1 = pd.Series(arr)
data2 = pd.Series(arr.cumsum()) # cumsum() = 누적합

# 데이터 출력
print(data1, data2)
```

## 기본 plt 조작법
  
### 선 그래프 그리기
  
plt.plot 함수는 데이터 시각화의 가장 기본적인 형태입니다. 주어진 데이터를 선 그래프로 그립니다. `plt.show()` 함수는 그래프를 화면에 출력합니다.

```python
plt.plot(data1)
plt.show()
```

### x축, y축 레이블 추가하기

`plt.xlabel`과 `plt.ylabel` 함수는 각각 x축과 y축에 라벨을 추가합니다. 이 함수들은 문자열을 인자로 받아 해당 축에 라벨을 표시합니다. 

```python
plt.plot(data1)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

### Title 만들기

`title` 함수는 그래프에 제목을 추가합니다. 문자열과 더불어 인자를 입력하면 아래와 같이 그 크기와 특징을 조절할 수 있습니다.

```python
plt.plot(data1)
plt.title("Grapth title", loc="center", pad=None)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

# Legend(범례)  

`plt.legend()` 함수는 그래프에 범례를 추가합니다. `loc` 인자는 범례의 위치를 지정하며, 가능한 값으로는 'upper left', 'upper right', 'lower left', 'lower right' 등이 있습니다. 기본값은 'best'로, 그래프의 비어 있는 부분에 자동으로 위치합니다.  

`label` 인자는 데이터 시리즈에 라벨을 추가합니다. 이렇게 하면 그래프에 여러 데이터 시리즈가 있을 때 각각을 구분하기 쉽습니다. 

```python
plt.plot(data1, label='Data 1')
plt.plot(data2, label='Data 2')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='upper left')
plt.show()
```

### 눈금 설정
  
x축과 y축의 눈금을 설정하여 그래프를 더욱 세밀하게 조정할 수 있습니다. `plt.xticks`와 `plt.yticks` 함수는 각각 x축과 y축의 눈금을 설정합니다. `ticks` 인자는 눈금의 위치를 지정하며, `np.arange`와 `np.linspace` 함수를 이용해 범위를 설정할 수 있습니다. `np.arange(0, 20, 2)`는 0에서 20까지 2 간격으로 눈금을 설정하고, `np.linspace(0, 1, 5)`는 0에서 1까지 5개의 눈금을 균등하게 설정합니다.



```python
plt.plot(data1, label='Data 1')
plt.plot(data2, label='Data 2')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')

plt.xticks(ticks=np.arange(0, 20, 1))
plt.yticks(ticks=np.linspace(0, max(data2), 10))

plt.legend(loc='upper left')
plt.show()
```

### 그래프 커스터마이징

#### marker 다르게 설정하는 법

`marker` 인자는 데이터 포인트를 표시하는 마커의 모양을 지정합니다. 가능한 값으로는 'o' (원), 's' (사각형), 'D' (다이아몬드), '^' (위쪽 삼각형), 'v' (아래쪽 삼각형), '>' (오른쪽 삼각형), '<' (왼쪽 삼각형) 등이 있습니다. 이는 데이터 포인트를 시각적으로 강조하는 데 사용됩니다.

[공식 문서](https://matplotlib.org/stable/api/markers_api.html)

```python
plt.plot(data1, marker='o')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

#### 선 색상 다르게 설정하는 법

`color` 인자는 선의 색상을 지정합니다. 가능한 값으로는 'red', 'blue', 'green', 'black', 'yellow' 등의 색상 이름을 사용할 수 있으며, '#RRGGBB' 형태의 HTML 컬러 코드도 사용할 수 있습니다. 이는 그래프의 시각적 구별을 쉽게 만들어줍니다.

[공식 문서](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
plt.plot(data1, color='red')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

#### 선 종류 다르게 설정하는 법

`linestyle` 인자는 선의 종류를 지정합니다. 가능한 값으로는 '-' (실선), '--' (점선), '-.' (점선과 실선 혼합), ':' (점선) 등이 있습니다. 이는 그래프의 시각적 스타일을 다양하게 만드는 데 사용됩니다.

[공식 문서](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
plt.plot(data1, linestyle='--')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

```python
plt.plot(data1, marker='o', color='green', linestyle='--')

# 'o' (원), 's' (사각형), 'D' (다이아몬드), '^' (위쪽 삼각형), 'v' (아래쪽 삼각형), '>' (오른쪽 삼각형), '<' (왼쪽 삼각형) 등
# 'red', 'blue', 'green', 'black', 'yellow' 등 or '#RRGGBB' 형태의 HTML 컬러 코드도 사용
#'-' (실선), '--' (점선), '-.' (점선과 실선 혼합), ':' (점선) 등

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

### 그래프의 범위 제한

`plt.xlim`과 `plt.ylim` 함수는 각각 x축과 y축의 표시 범위를 설정합니다. 인자로 최소값과 최대값을 지정하며, 이는 그래프에서 표시되는 데이터의 범위를 조정하여 보다 명확하게 특정 부분을 강조할 수 있습니다.

```python
plt.plot(data1)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.xlim(0, 10)
plt.ylim(0, 1)
plt.show()
```

### Figure와 Axes

Matplotlib가 그리는 그림은 Figure 객체, Axes 객체, Axis 객체 등으로 구성됩니다.  

- Figure: 전체 그림을 나타내는 캔버스나 종이입니다. 그림의 전체적인 레이아웃과 스타일을 정의합니다.

- Axes: 하나의 플롯을 나타내며, Figure 안에 여러 개의 Axes가 포함될 수 있습니다. 각 Axes는 개별적인 플롯을 의미합니다.

- Axis: 가로축이나 세로축을 나타냅니다. 축의 라벨, 눈금, 범위를 정의합니다.

Figure와 Axes는 `plt.figure`와 `plt.subplot` 명령을 사용하여 생성할 수 있습니다. Figure 객체는 `matplotlib.figure.Figure` 클래스 객체로, 하나의 Figure는 하나 이상의 Axes 객체를 포함합니다. Axes 객체는 `matplotlib.axes.Axes` 클래스 객체로, 두 개 이상의 Axis 객체를 포함합니다.

```python
# 데이터 생성
data1 = np.random.randn(100)
data2 = np.random.randn(100)
data3 = np.random.randn(100)
data4 = np.random.randn(100)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))

ax[0, 0].plot(data1)
ax[0, 0].set_title('data1')

ax[0, 1].plot(data2)
ax[0, 1].set_title('data2')

ax[1, 0].plot(data3)
ax[1, 0].set_title('data3')

ax[1, 1].plot(data4)
ax[1, 1].set_title('data4')
```

### 서브플롯

서브 플롯을 사용하면 하나의 창에 여러 그래프를 동시에 표시할 수 있어, 데이터의 여러 측면을 한눈에 비교할 수 있습니다.

Matplotlib에서 서브 플롯을 생성하는 기본적인 방법은 `plt.subplot()`함수를 사용하는 것입니다. 이 함수는 그리드 형태의 Axes 객체들을 생성합니다. Figure가 행렬이고 Axes가 행렬의 원소라고 생각하면 됩니다. 예를 들어, 위와 아래 두 개의 플롯이 있는 경우 행이 2이고 열이 1인 2x1 행렬입니다. subplot 명령은 세 개의 인수를 가지며, 처음 두 개의 원소가 전체 그리드 행렬의 모양을 지시하고, 세 번째 인수가 어느 플롯인지 지정합니다. 첫 번째 플롯은 subplot`(2, 1, 1)`로, 두 번째 플롯은 subplot`(2, 1, 2)`로 지정합니다.

```python
# 2x2 서브 플롯 생성
plt.figure(figsize=(10, 8))

# 
plt.plot(data1, label='Data 1')
plt.title("First Plot")

# 
plt.plot(data2, label='Data 2')
plt.title("Second Plot")

# 
plt.plot(data3, label='Data 3')
plt.title("Third Plot")

# 
plt.plot(data4, label='Data 4')
plt.title("Fourth Plot")

plt.tight_layout()
plt.show()
```

## 연습문제

### 실습 문제 1: 서브 플롯 생성하기

1. 2x1 그리드의 서브 플롯을 생성합니다.
2. 첫 번째 플롯에는 np.sin 함수를 사용하여 0부터 10까지의 사인(sin) 그래프를 그립니다.
3. 두 번째 플롯에는 np.cos 함수를 사용하여 0부터 10까지의 코사인(cos) 그래프를 그립니다.
4. 각 그래프에 적절한 제목을 추가합니다 ('Sine Wave', 'Cosine Wave').
5. x축에는 'Time'이라는 라벨을, y축에는 'Amplitude'라는 라벨을 각각 추가합니다.

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성

t = np.linspace(0, 10, 50)
sin_wave = np.sin(t)
cos_wave = np.cos(t)

# 서브 플롯 생성

fig, ax = plt.subplots(2, 1, figsize=(10, 8))

ax[0].plot(t, sin_wave)
ax[0].set_title('Sine Wave')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')

ax[1].plot(t, cos_wave)
ax[1].set_title('Cosine Wave')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Amplitude')

plt.tight_layout()

```

### 실습 문제 2: 서로 다른 스타일의 그래프를 서브 플롯에서 사용하기

1. 1x2 그리드의 서브 플롯을 생성합니다.
2. 첫 번째 플롯에는 무작위 데이터 'data1'을 생성하여 빨간색 점선 그래프를 그립니다.
3. 두 번째 플롯에는 무작위 데이터  'data2'를 생성하여 파란색 실선 그래프를 그립니다.

5. 두 플롯 모두 x축에는 'Index', y축에는 'Value'라는 라벨을 추가합니다.


```python
# 데이터 생성
data1 = np.random.randn(50).cumsum()
data2 = np.random.randn(50).cumsum()

# 서브 플롯 생성
#

fig, ax = plt.subplots(1, 2, figsize=(10, 8))

ax[0].plot(data1, linestyle='--', color='red')
ax[0].set_title('Red Dashed Line')
ax[0].set_xlabel('Index')
ax[0].set_ylabel('Value')

ax[1].plot(data2)
ax[1].set_title('Blue Solid Line')
ax[1].set_xlabel('Index')
ax[1].set_ylabel('Value')

plt.tight_layout()
```
