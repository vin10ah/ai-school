### 원형 그래프(Pie chart)

원형 그래프는 범주형 데이터의 비율을 시각화하는 데 주로 사용됩니다. 각 카테고리가 전체에서 차지하는 비율을 직관적으로 표현할 수 있기 때문에, 설문 조사 결과나 인구 분포 같은 데이터를 나타낼 때 효과적입니다.

가장 기본적인 원형 그래프를 그리는 방법은 plt.pie() 함수를 사용하는 것입니다. 데이터의 각 부분이 전체에서 차지하는 비율을 자동으로 계산하여 그래프에 표시합니다.

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 생성
data = pd.Series([215, 130, 245, 210], index=["Apples", "Bananas", "Cherries", "Dates"])
explode = (0.1, 0, 0, 0)

# 원형 그래프 그리기
plt.figure(figsize=(6, 6))
plt.pie(data, autopct='%1.1f%%', labels=data.index, explode=explode)
plt.title('Fruit Consumption Pie Chart')

plt.legend(data.index)
plt.show()
```

이 코드는 네 가지 유형의 과일(Apples, Bananas, Cherries, Dates)의 소비량을 나타내는 원형 그래프를 생성합니다. `autopct` 파라미터는 각 범주가 차지하는 비율을 퍼센트 형태로 표시합니다.

### 막대 그래프(Bar Chart)
막대 그래프는 범주형 데이터를 수량화된 막대 형태로 나타내어 각 카테고리 간 비교를 용이하게 합니다. 막대 그래프는 세로형(수직) 막대 그래프와 가로형(수평) 막대 그래프가 있으며, 매출 데이터, 설문 조사 결과, 인구 통계 등 다양한 데이터를 비교할 때 널리 사용됩니다.
  
기본 막대 그래프를 그리기 위해 `plt.bar()` 함수를 사용할 수 있습니다. 각 막대는 카테고리를 대표하며, 막대의 높이는 해당 카테고리의 값(빈도, 수량 등)을 나타냅니다.

```python
# 데이터 생성
categories = ['Red', 'Blue', 'Green', 'Yellow', 'Black']
values = [10, 15, 7, 5, 10]

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(categories, values)
plt.xlabel('Colors')
plt.ylabel('Values')
plt.title('Color Frequency Bar Chart')
plt.show()
```

막대 그래프를 그릴 때 주로 사용하는 몇 가지 인자들을 소개합니다:

- `color`: 막대의 색상을 설정합니다.
- `edgecolor`: 막대의 테두리 색상을 설정합니다.
- `width`: 막대의 너비를 설정합니다.
- `align`: 막대의 정렬 방식을 설정합니다. 기본값은 'center'입니다.

다음은 이러한 인자들을 사용하여 막대 그래프를 커스터마이징한 예제입니다.


```python
# 막대 그래프 그리기 - 커스터마이징
plt.figure(figsize=(12, 8))

# 기본 막대 그래프
plt.subplot(2, 2, 1)
plt.bar(categories, values)
plt.title('Default Bar Chart')
plt.xlabel('Colors')
plt.ylabel('Values')

# 테두리 색상 설정
plt.subplot(2, 2, 2)
plt.bar(categories, values, edgecolor='red')
plt.title('Bar Chart with Black Edges')s
plt.xlabel('Colors')
plt.ylabel('Values')

# 너비 설정
plt.subplot(2, 2, 3)
plt.bar(categories, values, edgecolor='red', width=0.5)
plt.title('Bar Chart with Width=0.5')
plt.xlabel('Colors')
plt.ylabel('Values')

# 정렬 방식 설정
plt.subplot(2, 2, 4)
plt.bar(categories, values, edgecolor='red', width=0.5, align='edge')
plt.title('Bar Chart with Edge Alignment')
plt.xlabel('Colors')
plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

### 히스토그램(Histogram)

히스토그램은 연속형 데이터를 구간별로 나누어 그 빈도를 시각화하는 데 사용됩니다. 이는 데이터의 분포를 파악하는 데 유용하며, 데이터가 특정 구간에 집중되어 있는지, 또는 분산되어 있는지 등을 쉽게 확인할 수 있습니다.

히스토그램과 막대 그래프는 비슷해 보이지만, 몇 가지 중요한 차이점이 있습니다:

- 공통점: 둘 다 데이터를 막대 형태로 시각화하여 비교할 수 있습니다.
- 차이점:
  - 히스토그램은 연속형 데이터를 다루며, 데이터의 구간(bin)을 자동 또는 수동으로 설정하여 각 구간의 빈도를 나타냅니다.
  - 막대 그래프는 범주형 데이터를 다루며, 각 막대는 개별 범주를 나타내고 막대의 높이는 해당 범주의 값(빈도, 수량 등)을 나타냅니다.


```python
# 데이터 생성
data = np.random.randn(1000)  # 평균이 0이고 표준편차가 1인 정규분포를 따르는 난수 1000개 생성

# 기본 히스토그램 그리기
plt.figure(figsize=(10, 6))
plt.hist(data)
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

히스토그램을 그릴 때 주로 사용하는 몇 가지 인자들을 소개합니다:

- `bins`: 히스토그램의 구간 수를 설정합니다. 기본값은 자동으로 설정됩니다.
- `alpha`: 막대의 투명도를 설정합니다. 0에서 1 사이의 값을 가지며, 0은 완전히 투명, 1은 불투명합니다.
- `color`: 막대의 색상을 설정합니다.
- `edgecolor`: 막대의 테두리 색상을 설정합니다.


```python
# 히스토그램 그리기 - 커스터마이징
plt.figure(figsize=(12, 8))

# 기본 히스토그램
plt.subplot(2, 2, 1)
plt.hist(data)
plt.title('Default Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 구간 수 설정
plt.subplot(2, 2, 2)
plt.hist(data, bins=50)
plt.title('Histogram with 50 Bins')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 투명도 설정
plt.subplot(2, 2, 3)
plt.hist(data, bins=30, alpha=0.3, color='blue')
plt.title('Histogram with Alpha=0.3')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 색상과 테두리 색상 설정
plt.subplot(2, 2, 4)
plt.hist(data, bins=30, color='red', edgecolor='yellow')
plt.title('Histogram with Red Bars and Yellow Edges')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### 박스 그래프(Box Plot)

박스 그래프는 데이터의 분포와 이상치를 시각화하는 데 사용됩니다. 데이터의 최소값, 1사분위수(Q1), 중앙값(Q2), 3사분위수(Q3), 최대값을 한눈에 확인할 수 있으며, 이상치(outliers)도 쉽게 파악할 수 있습니다.

박스 그래프는 주로 다음과 같은 유형의 데이터를 다룰 때 사용됩니다:
- 데이터의 분위수를 중심으로 분포를 비교할 때
- 이상치를 식별할 때
- 여러 그룹 간의 데이터 분포를 비교할 때

```python
# 데이터 생성
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# 기본 박스 그래프 그리기
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title('Basic Box Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

### 산점도(Scatter Plot)

산점도는 두 연속형 변수 간의 관계를 시각화하는 데 사용됩니다. 각 점은 두 변수의 값을 나타내며, 두 변수 간의 상관관계, 패턴, 분포 등을 쉽게 확인할 수 있습니다.

산점도는 주로 다음과 같은 유형의 데이터를 다룰 때 사용됩니다:
- 두 변수 간의 상관관계를 분석할 때
- 데이터의 분포와 패턴을 시각화할 때
- 이상치(outliers)를 식별할 때

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + np.random.randn(100) * 0.1  # x와 y의 관계를 정의

# 기본 산점도 그리기
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.title('Basic Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

산점도를 그릴 때 주로 사용하는 몇 가지 인자들을 소개합니다:

- `c`: 각 점의 색상을 설정합니다.
- `s`: 각 점의 크기를 설정합니다.
- `alpha`: 점의 투명도를 설정합니다. 0에서 1 사이의 값을 가지며, 0은 완전히 투명, 1은 불투명합니다.
- `marker`: 점의 모양을 설정합니다.

다음은 이러한 인자들을 사용하여 산점도를 커스터마이징한 예제입니다.

```python
# 산점도 그리기 - 커스터마이징
plt.figure(figsize=(12, 8))

# 기본 산점도
plt.subplot(2, 2, 1)
plt.scatter(x, y)
plt.title('Default Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 색상 설정
plt.subplot(2, 2, 2)
plt.scatter(x, y, c='red')
plt.title('Scatter Plot with Red Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 크기 설정
plt.subplot(2, 2, 3)
plt.scatter(x, y, s=100)
plt.title('Scatter Plot with Larger Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 색상, 크기, 투명도, 모양 설정
plt.subplot(2, 2, 4)
plt.scatter(x, y, c='green', s=30, alpha=0.3, marker='^')
plt.title('Scatter Plot with Custom Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

    - y는 3 * x + 5에 약간의 노이즈를 더한 값입니다.

3. 점의 색상을 파란색으로 설정한 산점도를 그립니다.
4. 점의 크기를 70으로 설정한 산점도를 그립니다.
5. 점의 색상, 크기, 투명도, 모양을 커스터마이징한 산점도를 그립니다.

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
np.random.seed(42)
x = np.random.uniform(0, 10, 100)
y = (3 * x + 5) * 0.03

# 산점도 그리기
plt.figure(figsize=(12, 8))

# 기본 산점도
plt.subplot(2, 2, 1)
plt.scatter(x, y)
plt.title('Default Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 색상 설정
plt.subplot(2, 2, 2)
plt.scatter(x, y, c='blue')
plt.title('Scatter Plot with Blue Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 크기 설정
plt.subplot(2, 2, 3)
plt.scatter(x, y, s=70)
plt.title('Scatter Plot with Larger Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 색상, 크기, 투명도, 모양 설정
plt.subplot(2, 2, 4)
plt.scatter(x, y, c='yellow', s=50, marker='*', alpha=0.5)
plt.title('Scatter Plot with Custom Dots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```
