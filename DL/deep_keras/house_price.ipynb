```python
import numpy as np
import pandas as pd

df = pd.read_csv('./house_price.csv')
```

```python
df.dropna(axis=1, how='any', inplace=True)
df.shape
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
# data type이 object인 데이터들 label encoding 

col_lst = [col for col in df.columns if df[col].dtype == 'object']
col_lst
le = LabelEncoder()

for col in col_lst:
	df[col] = le.fit_transform(df[col])

df
```

```python
df.info()
```

```python
df.drop('Id', axis=1, inplace=True)
df
```

```python
train_data = df.iloc[:1200, :]
test_data = df.iloc[1200:, :]

len(train_data)
```

```python
X_train = df.drop('SalePrice', axis=1)
y_train = df['SalePrice']
```

```python
# X_train = train_data.drop('SalePrice', axis=1)
# y_train = train_data['SalePrice']
```

```python
def scaling(x_trn, x_tst, y_trn, y_tst, scaler):
	
	scaled_x_trn = scaler.fit_transform(x_trn)
	scaled_x_tst = scaler.transform(x_tst)

	scaled_y_trn = scaler.fit_transform(y_trn.values.reshape(-1,1)).flatten()
	scaled_y_tst = scaler.transform(y_tst.values.reshape(-1,1)).flatten()

	return scaled_x_trn, scaled_x_tst, scaled_y_trn, scaled_y_tst
```

```python
from tensorflow import keras
from keras import models
from keras.layers import Dense
from tensorflow.keras.utils import plot_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler


x_trn, x_tst, y_trn, y_tst = train_test_split(X_train, y_train, test_size=0.2, random_state=0)

scaler = StandardScaler()

scaled_x_trn, scaled_x_tst, scaled_y_trn, scaled_y_tst = scaling(x_trn, x_tst, y_trn, y_tst, scaler)

model = keras.Sequential([
	Dense(64, input_shape=(60,), activation='relu'),
	Dense(128, activation='relu'),
	Dense(128, activation='relu'),
	Dense(64, activation='relu'),
	Dense(1)
])

optimizer = keras.optimizers.Adam(learning_rate=0.001)

model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
model.fit(scaled_x_trn, scaled_y_trn, epochs=1000, batch_size=64, verbose=1)

test_mse_score, test_mae_score = model.evaluate(scaled_x_tst, scaled_y_tst)
print(test_mae_score)
```
